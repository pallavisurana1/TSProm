```python
import os
import json
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
# GENA-LM
base_path = "/data/projects/dna/pallavi/DNABERT_runs/DATA_RUN/gena-lm_finetune/unique_TSS/human_mouse"

# Expected subfolders
conditions = [
    "tsp_brain_low", "tsp_brain_null", "tsp_brain_tenh", "tsp_brain_wide",
    "tsp_liver_low", "tsp_liver_null", "tsp_liver_tenh", "tsp_liver_wide",
    "tsp_testis_low", "tsp_testis_null", "tsp_testis_tenh", "tsp_testis_wide"
]
sizes = ["2000", "3000", "4000"]

results = []

for condition in conditions:
    for size in sizes:
        size_dir = os.path.join(base_path, condition, size)
        if not os.path.isdir(size_dir):
            continue

        # lr subfolders
        for lr_dir in os.listdir(size_dir):
            run_path = os.path.join(size_dir, lr_dir)
            eval_path = os.path.join(run_path, "eval_results.json")

            entry = {
                "Condition": condition,
                "Size": size,
                "LR": lr_dir,
                "Precision": "",
                "Recall": "",
                "F1": "",
                "MCC": "",
                "Accuracy": "",
                "Loss": "",
                "Path": eval_path if os.path.exists(eval_path) else ""
            }

            if os.path.exists(eval_path):
                try:
                    with open(eval_path, "r") as f:
                        metrics = json.load(f)

                    entry.update({
                        "Precision": round(metrics.get("eval_precision", 0) * 100, 2),
                        "Recall": round(metrics.get("eval_recall", 0) * 100, 2),
                        "F1": round(metrics.get("eval_f1", 0) * 100, 2),
                        "MCC": round(metrics.get("eval_matthews_correlation", 0) * 100, 2),
                        "Accuracy": round(metrics.get("eval_accuracy", 0) * 100, 2),
                        "Loss": round(metrics.get("eval_loss", 0), 4) if "eval_loss" in metrics else ""
                    })
                except Exception as e:
                    print(f"Error reading {eval_path}: {e}")

            results.append(entry)

# DataFrame
df = pd.DataFrame(results)
df.to_csv('/data/private/psurana/TSpDNA2/src/files/GENA-LM_TSP_vs_TransTExRest.csv')
```

```python
# for dna2
base_path = "/data/projects/dna/pallavi/DNABERT_runs/DATA_RUN/gena-lm_finetune/unique_TSS/human_mouse"

# Expected subfolders
conditions = [
    "tsp_brain_low", "tsp_brain_null", "tsp_brain_tenh", "tsp_brain_wide",
    "tsp_liver_low", "tsp_liver_null", "tsp_liver_tenh", "tsp_liver_wide",
    "tsp_testis_low", "tsp_testis_null", "tsp_testis_tenh", "tsp_testis_wide"
]

sizes = ["2000", "3000", "4000"]

results = []

for condition in conditions:
    for size in sizes:
        run_base = os.path.join(base_path, condition, size)
        if not os.path.exists(run_base):
            continue

        for run_name in os.listdir(run_base):
            run_path = os.path.join(run_base, run_name)
            results_dir = os.path.join(run_path, "results")

            if not os.path.isdir(results_dir):
                continue

            for run_subdir in os.listdir(results_dir):
                eval_path = os.path.join(results_dir, run_subdir, "eval_results.json")

                # Default entry with empty values
                entry = {
                    "Condition": condition,
                    "Size": size,
                    "Experiment": run_name,
                    "Run": run_subdir,
                    "Precision": '',
                    "Recall": '',
                    "F1": '',
                    "MCC": '',
                    "Accuracy": '',
                    "Path": eval_path if os.path.exists(eval_path) else ''
                }

                if os.path.exists(eval_path):
                    try:
                        with open(eval_path, 'r') as f:
                            metrics = json.load(f)

                        entry.update({
                            "Precision": round(metrics.get('eval_precision', 0) * 100, 2),
                            "Recall": round(metrics.get('eval_recall', 0) * 100, 2),
                            "F1": round(metrics.get('eval_f1', 0) * 100, 2),
                            "MCC": round(metrics.get('eval_matthews_correlation', 0) * 100, 2),
                            "Accuracy": round(metrics.get('eval_accuracy', 0) * 100, 2),
                        })
                    except Exception as e:
                        print(f"Error reading {eval_path}: {e}")

                results.append(entry)

# Convert to DataFrame
df = pd.DataFrame(results)

# Show first few rows
df.head(2)
```

```python
dna2=pd.read_csv('/data/private/psurana/TSpDNA2/src/files/DNABERT2_TSP_vs_TransTExRest.csv')
genaLM = pd.read_csv('/data/private/psurana/TSpDNA2/src/files/GENA-LM_TSP_vs_TransTExRest.csv')
```

```python
def pick_best_models(df):
    # make sure metrics are numeric
    for c in ["Precision", "Recall", "F1", "MCC", "Accuracy"]:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    # split Condition into Organ / Setting
    df = df.copy()
    df[["Organ", "Setting"]] = (
        df["Condition"]
        .str.replace(r"^tsp_", "", regex=True)
        .str.rsplit("_", n=1, expand=True)
    )

    # sort so best metrics come first
    df_sorted = df.sort_values(
        ["Organ", "Size", "F1", "MCC", "Accuracy"],
        ascending=[True, True, False, False, False]
    )

    # pick best row per (Organ, Size)
    best = df_sorted.groupby(["Organ", "Size"], as_index=False).first()
    return best


best_modelsDNA2 = pick_best_models(dna2)
best_modelsgenaLM = pick_best_models(genaLM)
```

```python
print(best_modelsDNA2.shape, best_modelsgenaLM.shape)
```

```python
dna2_ = best_modelsDNA2.copy()
gena_ = best_modelsgenaLM.copy()
dna2_["Model"] = "DNABERT2"
gena_["Model"] = "GenaLM"
df = pd.concat([dna2_, gena_], ignore_index=True)
df.head(2)
```

```python
df
```

```python

#  collapse to best per Organ+Size+Model (by F1, then MCC, Accuracy) 
df_best = (
    df.sort_values(["F1", "MCC", "Accuracy"], ascending=[False, False, False])
      .groupby(["Organ", "Size", "Model"], as_index=False)
      .first()
)

metrics = ["F1", "MCC", "Accuracy"]
colors = {"F1": "purple", "MCC": "orange", "Accuracy": "blue"}
models = ["DNABERT2", "GenaLM"]
linestyles = {"DNABERT2": "-", "GenaLM": "--"}
markers = {"DNABERT2": "o", "GenaLM": "s"}

organs = ["brain", "liver", "testis"]
fig, axes = plt.subplots(1, len(organs), figsize=(10, 5), sharey=True)

for ax, organ in zip(axes, organs):
    organ_df = df_best[df_best["Organ"] == organ]
    if organ_df.empty:
        ax.set_title(f"{organ.capitalize()}: (no data)", fontsize=16, fontweight="bold")
        ax.set_xticks([])
        continue

    # ensure sorted sizes
    organ_df = organ_df.sort_values(["Size", "Model"])
    size_labels = organ_df["Size"].unique().tolist()
    size_map = {s: i for i, s in enumerate(size_labels)}

    for metric in metrics:
        for model in models:
            sub = organ_df[organ_df["Model"] == model]
            if sub.empty:
                continue
            xvals = [size_map[s] for s in sub["Size"]]

            # highlight DNABERT2
            lw = 3 if model == "DNABERT2" else 2
            ms = 9 if model == "DNABERT2" else 6

            ax.plot(
                xvals, sub[metric],
                linestyle=linestyles[model],
                marker=markers[model],
                markersize=ms,
                linewidth=lw,
                label=f"{metric} ({model})",  # only used for legend construction
                color=colors[metric]
            )

    # formatting
    ax.set_title(f"{organ.capitalize()}", fontsize=20, fontweight="bold")
    ax.set_xlabel("Sequence length", fontsize=16, fontweight="bold")
    ax.set_xticks(range(len(size_labels)))
    ax.set_xticklabels(size_labels, fontsize=14, fontweight="bold")
    ax.set_ylim(0, 100)
    ax.grid(True, alpha=0.3)
    if ax == axes[0]:
        ax.set_ylabel("Metric (%)", fontsize=16, fontweight="bold")
    ax.tick_params(axis="y", labelsize=14)
    ax.tick_params(axis="x", labelsize=14)

#  Legends: split into two clean rows 
metric_handles = [Line2D([0], [0], color=colors[m], lw=3, label=m) for m in metrics]
model_handles  = [Line2D([0], [0], color="black", lw=3, linestyle=linestyles[m],
                         marker=markers[m], markersize=8, label=m) for m in models]

fig.legend(handles=model_handles,
           loc="upper center", ncol=len(models),
           title="Models", bbox_to_anchor=(0.5, 1.20),
           fontsize=12, title_fontsize=14)
fig.legend(handles=metric_handles,
           loc="upper center", ncol=len(metrics),
           title="Metrics", bbox_to_anchor=(0.5, 1.08),
           fontsize=12, title_fontsize=14)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("/data/private/psurana/TSpDNA2/src/files/coomparison_methods.svg", format="svg", dpi=1200, bbox_inches="tight")
plt.show()
```
